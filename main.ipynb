```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import math
```

Making a uniformly random list of x and y values

```python
count = 1000

xs = np.random.uniform(-1, 1, count)
ys = np.random.uniform(-1, 1, count)

print(xs)
print(ys)
```

```python
print(type(xs))
```

Setting the colors inside the circle to be green and outside to be red

```python
colors = []
inside = 0
outside = 0

for x, y in zip(xs, ys):
    if x**2 + y**2 <= 1:
        colors.append('green')
        inside += 1
    else:
        colors.append('red')
        outside += 1
```

```python
plt.scatter(xs, ys, c=colors)
plt.grid()
circle = Circle((0, 0), 1, color='blue', fill=False)
ax = plt.gca()
ax.add_patch(circle)
plt.show()
```

```python
print(f"Number of marbles inside the circle: {inside}")
print(f"Number of marbles outside the circle: {outside}")
print(f"Total number of marbles: {inside + outside}")
print(f"Ratio of inside to total: {inside/(inside + outside)}")
print(f"Ratio of pi/4: {math.pi/4}")
```

The area of the square where the marbles can fall is 4 since the side lengths are 2.

The radius of the circle is 1 so the area is $\pi$

So the ratio of area of the circle to the area of the entire square would be $\frac{\pi}{4}$

So we can see it as $\frac{\text{the number of marbles inside the circle (green)}}{\text{the number of marbles inside the entire square (green + red)}} = \frac{786}{1000} = 0.786$

We also see that $\frac{\pi}{4} \approx 0.785$

We can see that these values are very close. We can make them even closer by increasing the number of samples or increasing the value of `count`. Currently `count = 1000`

So we can say that $\frac{\text{the number of marbles inside the circle}}{\text{the number of marbles inside the entire square}} \approx \frac{\text{the area of the circle}}{\text{the area of the square}}$
